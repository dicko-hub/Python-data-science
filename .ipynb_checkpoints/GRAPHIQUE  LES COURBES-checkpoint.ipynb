### Présentation du dataset

```python
import pandas as pd
```

```python
unrate = pd.read_csv("unrate.csv")
```

```python
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
unrate[0:12]
```

### Introduction à matplotlib

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot()
plt.show()
```

###  Ajouter des données à nos graphiques

```python
unrate
```

```python
unrate_1948 = unrate[(unrate["DATE"] < "1949-01-01") & (unrate["DATE"] > "1947-12-31")]
unrate_1948
```

```python
plt.plot(unrate_1948["DATE"],unrate_1948["UNRATE"])
plt.show
```

### Créer une rotation sur les données de l'axe x

```python
# plt.xsticks()
#matplotlib.pyplot.xsticks(*args, **kwargs)
```

```python
plt.plot(unrate_1948["DATE"],unrate_1948["UNRATE"])
plt.xticks(rotation = 45)
plt.show
```

### Ajouter les intitulés des axes et un titre

```python
# plt.xlabel(), plt.ylabel(), plt.title()
plt.plot(unrate_1948["DATE"],unrate_1948["UNRATE"])
plt.xticks(rotation = 35)
plt.xlabel("Mois")
plt.ylabel("Taux de chômage (%)")
plt.title("Evolution du taux de chômage en 1948 au USA")
plt.show
```

###  Ajouter des données aux graphiques

```python
# graphe 1948 et 1949 sur le chomage
fig = plt.figure()
axe1 = fig.add_subplot(2,1,1)
axe2 = fig.add_subplot(2,1,2)

axe1.plot(unrate["DATE"][0:12], unrate["UNRATE"][0:12])
axe2.plot(unrate["DATE"][12:24], unrate["UNRATE"][12:24])

plt.show()
```

###  Améliorer l'affichage

```python
# graphe 1948 et 1949 sur le chomage
fig = plt.figure(figsize = (15,8))
axe1 = fig.add_subplot(2,1,1)
axe2 = fig.add_subplot(2,1,2)

axe1.plot(unrate["DATE"][0:12], unrate["UNRATE"][0:12])
axe2.plot(unrate["DATE"][12:24], unrate["UNRATE"][12:24])

axe1.set_title("Taux de chômage en 1948")
axe2.set_title("Taux de chômage en 1949")

plt.show()
```

### Comparaison de graphiques

```python
fig = plt.figure(figsize = (12,12))

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset["DATE"],subset["UNRATE"])
    
plt.show()
```

###  Superposition des courbes

```python
# pandas.Series.dt
unrate["MONTH"] = unrate["DATE"].dt.month
```

```python
fig = plt.figure()

plt.plot(unrate[0:12]["MONTH"],unrate[0:12]["UNRATE"])
plt.plot(unrate[12:24]["MONTH"],unrate[12:24]["UNRATE"])

plt.show()
```

```python
fig = plt.figure(figsize=(6,3))

plt.plot(unrate[0:12]["MONTH"],unrate[0:12]["UNRATE"], c="red")
plt.plot(unrate[12:24]["MONTH"],unrate[12:24]["UNRATE"], c="blue")

plt.show()
```

###  Ajouter plus de courbes

```python
fig = plt.figure(figsize=(10,6))

plt.plot(unrate[0:12]["MONTH"],unrate[0:12]["UNRATE"], c="red")
plt.plot(unrate[12:24]["MONTH"],unrate[12:24]["UNRATE"], c="blue")
plt.plot(unrate[24:36]["MONTH"],unrate[24:36]["UNRATE"], c="green")
plt.plot(unrate[36:48]["MONTH"],unrate[36:48]["UNRATE"], c="orange")
plt.plot(unrate[48:60]["MONTH"],unrate[48:60]["UNRATE"], c="black")

plt.show()
```

```python
# autre cas
fig = plt.figure(figsize = (10,6))
colors = ["red","blue","green","orange","black"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset["MONTH"],subset["UNRATE"], c = colors[i])
    
plt.show()
```

###  Ajouter une légende

```python
# plt.legend() : pour le tout ou axes.legend() : pour un axe en particulier
fig = plt.figure(figsize = (10,6))
colors = ["red","blue","green","orange","black"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset["MONTH"],subset["UNRATE"], c = colors[i], label = label)
    
plt.legend(loc="upper left") 
plt.show()
```

### Courbe finale

```python
# plt.title(), plt.xlabel(), plt.ylabel()
fig = plt.figure(figsize = (10,6))
colors = ["red","blue","green","orange","black"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset["MONTH"],subset["UNRATE"], c = colors[i], label = label)
    
plt.legend(loc="upper left") 
plt.title("Evolution du taux de chômage de 1948 à 1952")
plt.xlabel("Mois (de 1 à 12)")
plt.ylabel("Taux de chômage (en %)")
plt.show()
```
