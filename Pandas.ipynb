### Cas pratique

```python
list_col_names = food_info.columns.tolist()

gram_columns = []

for name in list_col_names:
    if (name.endswith('(g)')):
        gram_columns.append(name)
        
gram_df = food_info[gram_columns]
gram_df.head(3)
```
