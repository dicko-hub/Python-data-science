```python
import pandas as pd

food_info = pd.read_csv('food_info.csv')
```

```python
food_info.head()
```

### Transformer une colonne

```python
soduim_grams = food_info["Sodium_(mg)"]/1000
sugar_miligrams = food_info["Sugar_Tot_(g)"]*1000
sugar_miligrams
```

### Opérations mathématiques entre colonnes

```python
grams_of_per_gram_of_water = food_info["Protein_(g)"] / food_info["Water_(g)"]
miligrams_of_calcium_and_iron = food_info["Calcium_(mg)"] + food_info["Iron_(mg)"]
grams_of_per_gram_of_water
```

### Créer un indice nutritionnel

```python
# Score = 2*(Protein(g)) - 0.75*(LipidTot(g))
var_protein = 2 * food_info["Protein_(g)"]

var_lipid = -0.75 * food_info["Lipid_Tot_(g)"]

rating = var_protein + var_lipid

rating.head()
```

### Normaliser une colonne

```python
# series.max()
max_protein = food_info["Protein_(g)"].max()
normalized_protein = food_info["Protein_(g)"] / max_protein

max_lipid = food_info["Lipid_Tot_(g)"].max()
normalized_lipid = food_info["Lipid_Tot_(g)"] / max_lipid

normalized_protein.head()
```

### Créer une nouvelle colonne

```python
food_info['Normalized_Protein'] = normalized_protein
food_info['Normalized_Fat'] = normalized_lipid

food_info.head()
```

###  Créer une nouvelle colonne pour l'indice nutritionnel normalisé

```python
# Score = 2*(Protein(g)) - 0.75*(LipidTot(g))
food_info["Norm_Nutr_Index"] = 2*food_info["Normalized_Protein"] - 0.75*food_info["Normalized_Fat"]

food_info.head()
```

### Trier un dataframe

```python
# sort_values(col) : remplace pas l'originale
# sort_values(col, inplace = True, ascending = False ) : remplace l'originale
food_info.sort_values("Norm_Nutr_Index", inplace = True, ascending = False)
food_info.head()
```
